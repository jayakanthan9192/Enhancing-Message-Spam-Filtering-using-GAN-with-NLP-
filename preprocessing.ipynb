## Data loading

### Subtask:
Load the dataset "spam (2).csv" into a pandas DataFrame.


**Reasoning**:
Load the dataset "spam (2).csv" into a pandas DataFrame and display its first few rows.



```python
import pandas as pd

try:
    df = pd.read_csv('spam (2).csv', encoding='latin-1')
    display(df.head())
    display(df.info())
except FileNotFoundError:
    print("Error: 'spam (2).csv' not found. Please ensure the file exists in the current directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the loaded dataset to understand its structure, identify potential issues, and inform subsequent cleaning and preparation steps.


**Reasoning**:
Analyze the distribution of features, identify potential inconsistencies or irrelevant columns, and summarize the findings.



```python
# Analyze the distribution of features
print(df['v1'].value_counts())  # Frequency of categories in 'v1'
print(df['v2'].value_counts().head(10)) # Top 10 most frequent messages in 'v2'

# Identify potential inconsistencies or irrelevant columns
print(df.isnull().sum()) # Check for missing values again

# Analyze 'Unnamed' columns
for col in ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']:
    print(f"\nAnalysis of column '{col}':")
    print(df[col].value_counts(dropna=False)) # Include NaN counts

# Summarize findings
print("\nSummary of Findings:")
print("1. Data Types and Missing Values:")
print(" - 'v1' and 'v2' have no missing values.")
print(" - 'Unnamed: 2', 'Unnamed: 3', and 'Unnamed: 4' have a large number of missing values.")
print(" - All columns are of object dtype, which might need further investigation, especially for 'v2'.")

print("\n2. Distribution of Features:")
print(" - 'v1' column shows the distribution of 'ham' and 'spam' messages.")
print(" - 'v2' column shows a wide variety of text messages, with some messages appearing multiple times.")

print("\n3. Potential Inconsistencies or Irrelevant Columns:")
print(" - 'Unnamed: 2', 'Unnamed: 3', and 'Unnamed: 4' contain mostly null values and appear irrelevant for the task.")
print(" - The 'v2' column should be examined for inconsistencies in text format.")

print("\n4. Columns with mostly null values:")
print(" - 'Unnamed: 2', 'Unnamed: 3', and 'Unnamed: 4' columns are candidates for removal due to high null value counts.")
print(" - The usefulness of these columns is questionable and they might be dropped in the next steps.")
```

## Data cleaning

### Subtask:
Clean the DataFrame `df` by removing irrelevant columns and handling duplicate rows.


**Reasoning**:
Drop the irrelevant columns and remove duplicate rows from the DataFrame `df` as instructed, then display the first 5 rows and the info of the cleaned DataFrame.



```python
# Drop irrelevant columns
df = df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])

# Remove duplicate rows
df = df.drop_duplicates(keep='first')

# Display first 5 rows and info
display(df.head())
display(df.info())
```

## Data preparation

### Subtask:
Convert the categorical feature 'v1' in the DataFrame `df` into numerical representations using label encoding.


**Reasoning**:
Convert the categorical feature 'v1' to numerical labels using LabelEncoder.



```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit the encoder on the 'v1' column
label_encoder.fit(df['v1'])

# Transform the 'v1' column and overwrite the original column
df['v1'] = label_encoder.transform(df['v1'])

# Display the first few rows and info to verify
display(df.head())
display(df.info())
```
